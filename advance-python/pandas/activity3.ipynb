### **Exercise**
- Add a new Column Revised_Salary, revised salary is 10% more than actual salary
- Then calculate total additional cost to company
- Which department faces the biggest impact

```python
# 1. Add a new Column Revised_Salary, revised salary is 10% more than actual salary
data["Revised_Salary"] = [salary+0.1*salary for salary in data["Salary"]]
data
```

```python
# 2. Calculate total additional cost to company
add_cost_to_company=data.groupby("Department")["Revised_Salary"].sum()-data.groupby("Department")["Salary"].sum()
print("Additinal Cost to company department wise:\n",add_cost_to_company)
print(f"\nTotal additional cost to company: {add_cost_to_company.sum()}")
```

```python
# 2. Which Department faces the biggest impact
biggest_impact = data.groupby("Department")["Revised_Salary"].sum()-data.groupby("Department")["Salary"].sum()
biggest_impact_id = biggest_impact.idxmax()
biggest_impact_value = biggest_impact.max()
print(f"The department facing the biggest impact is {biggest_impact_id} with an addtional cost of {biggest_impact_value}")
```
