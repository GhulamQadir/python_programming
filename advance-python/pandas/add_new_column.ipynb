```python
import pandas as pd

df = pd.read_excel('files/employee.xlsx')
```

```python
# adding new column in df2 manually (without any condition check)
df["bonus"] = [23000,120000,4366,23,43343,4343434,57567,7979,35454,545454,55454,787,3232,3232,434,32,12,2,]
df.head()
```

```python
# adding bonus
df["bonus"] = df["Salary"] * 0.1  # this method adds column at the end of the dataset
df.head()
```

```python
df.drop(columns=["bonus"], axis=1, inplace=True)
df.head()
```

```python
# adding bonus using insert method
# Syntax: df.insert(location, column_name, data)
df.insert(3, "bonus",df["Salary"] * 0.5)  # this method adds column at specific location
df.head(8)
```

```python
# adding bonus based on condition using list comprehension
df["Bonus1"] = [
        (
        50000 if ((des == "Manager") or (des == "Engineer")) else 
        40000 if (des == "Officer") else 
        30000 if (des == "Accountant") else 
        20000
        ) 
        for des in df["Designation"]
]
print(df)
```

```python
# Adding bonus based on Age using list comprehension
df["Bonus_Age"] = [
    (
        50000 if age >= 60 else
        40000 if age >= 50 else 
        30000 if age >= 40 else
        20000 if age >= 30 else 
        10000 if age > 20 else 5000
    )
    for age in df["Age"]
]
df.head(7)
```

```python
# adding bonus based on salary percentage
df["Bonus1"] = [
    (
        sal * 0.5 if sal >= 200000
        else sal * 0.25 if sal >= 150000 
        else sal * 0.15 if sal >= 75000 
        else sal * 0.1
    )
    for sal in df.Salary
]
df.head()
```

```python
# adding new column using function to increment salary
def bonus(sal):
    if sal > 20000:
        return sal * 0.5
    elif sal > 150000:
        return sal * 0.3
    elif sal >= 75000:
        return sal * 0.15
    else:
        return sal * 0.1


df["Bonus_func"] = df.Salary.apply(bonus)
df.head()
```

```python
# using lambda function
df["Bonus2"] = df.Salary.apply(
    lambda sal: (
        sal * 0.5 if sal >= 200000 
        else sal * 0.3 if sal >= 150000 
        else sal * 0.1
    )
)
df.head()
```

```python
# drop columns from dataframe
# axis = 1 for columns, 0 for rows
# inplace = True modifies the original DataFrame directly, instead of returning a new one.
df.drop(["bonus", "Bonus1","Bonus_Age","Bonus_func","Bonus2"], axis=1, inplace=True)
df.head()
```

```python
# adding bonus based on age using lambda function
"""axis=1 is used because we are applying the function row by row, and within each row, we can access multiple columns by their names."""
df["Bonus"] = df.apply(
    lambda row: (
        row["Salary"] * 0.5
        if row["Age"] >= 50
        else row["Salary"] * 0.4 if row["Age"] >= 40 else row["Salary"] * 0.1
    ),
    axis=1
)  
df.tail(8)
```
