```python
import pandas as pd


data = pd.read_excel("files/employee.xlsx")
data
```

```python
# filteration using boolean masking (single condition)
filtered_rows = data[data["Salary"] > 80000]
print("Employees with Salary greater than 80000")
filtered_rows
```

```python
data.describe()
```

```python
# filtering rows based on multiple conditions (single &(and))
filtered_rows = data[(data["Age"] < 30) & (data["Salary"] > 50000)]
print("Employees with age less than 30 AND Salary greater than 50000")
filtered_rows
```

```python
# filtering rows based on multiple conditions (single |(or))
filtered_rows = data[(data["Age"] > 40) | (data["Salary"] > 80000)]
print("Employees with age greater than 40 OR Salary greater than 80000")
filtered_rows
```

```python
# multiple conditions filteration
filtered_rows = data[(data["Age"] > 40) & ((data["Designation"] == "Manager") | (data["Designation"] == "Accountant"))]
print("Employees older than 40 AND designation is (Manager OR an Accountant)")
filtered_rows
```
