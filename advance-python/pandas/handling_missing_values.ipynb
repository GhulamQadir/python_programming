## Drop rows which have missing values

```python
# drop rows with missing values.
# axis = 0 for dropping rows and axis = 1 for dropping columns
df.dropna(axis=0, inplace=True) 
df
```

## Filling missing values in the rows

```python
df = pd.read_excel("files/employee.xlsx")
df
```

```python
# filling values one by one
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Department"].fillna(df["Department"].mode()[0], inplace=True)
df
```

```python
df = pd.read_excel("files/employee.xlsx")


# filling the missing values of columns at once
df.fillna({"Age": df["Age"].median(), 
           "Department":df["Department"].mode()[0], 
           "Salary":df["Salary"].mean()},  
          inplace=True) # insert first mode from the Department columns
df
```
