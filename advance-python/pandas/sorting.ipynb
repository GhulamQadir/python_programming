# **Sorting**

**Syntax:** data.sort_values(by = "Column_name", ascending = True/False, inplace = True)


```python
import pandas as pd

data = pd.read_excel("files/emp_bonus.xlsx")
data
```

```python
data.sort_values(by = "Salary", inplace=True)  # sorts in ascending order by default
print("Sorted Salary in ascending order")
data
```

```python
data.sort_values(by="Age", ascending=False, inplace=True)
print("Sorted Age in descending order")
data
```

## **Sort by multiple columns at once**
**Syntax:** data.sort_values(by = ["Column1","Column2"], ascending = [True, False], inplace = True)

```python
data.sort_values(by=["Age","Salary"], ascending=[True, False], inplace=True)
data
```
