## Update dataframe

```python
import pandas as pd

df = pd.read_excel("files/employee.xlsx")
df.head()
```

```python
# .loc is used to select and then update specific row or column value
# Syntax: data.loc[row_index, "Column name"] = new_value

# updating single value
df.loc[2, "Designation"] = "Accountant"
df
```

```python
# for updating whole column, use this method
# increasing salary by 5%
df["Salary"] = df["Salary"] * 1.5
df.head(7)
```
