```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel("../pandas/files/employee.xlsx")
print(data)
```

# **Now Visualizing Employees Data**


```python
# Bar Chart
freq = data.Designation.value_counts()
plt.bar(x=freq.index, height=freq.values, width=0.5)
```

```python
# Pie Chart
plt.pie(x=freq.values, labels=freq.index, autopct='%1.1f%%', shadow=True, wedgeprops={'linewidth': 5, 'edgecolor': 'grey'}
)
```
